```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
train = pd.read_csv("../input/train.csv")
test = pd.read_csv("../input/test.csv")
```

```python
train.head()
```

```python
train.shape
```

```python
train.isnull().sum()
```

```python
# Countplot for categorical columns
cat_col = ['term', 'grade', 'employment_length', 'purpose', 'application_type']

sns.set_theme()
fig, ax = plt.subplots(nrows=len(cat_col), ncols=2, figsize=(15, 50))
fig.subplots_adjust(hspace=.5)
plt.title('Countplot for categorical variables')

for i, col in enumerate(cat_col):
    sns.countplot(y=col, data=train, order=sorted(list(train[col].unique())), ax = ax[i][0])
    sns.countplot(y=col, data=test, order=sorted(list(test[col].unique())), ax = ax[i][1])
    ax[i][0].set_title('count of {}'.format(col))
    ax[i][1].set_title('count of {}'.format(col))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train, train['loan_status'], test_size=0.2, random_state=0, stratify=train['loan_status'])
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.25, random_state=0, stratify=y_train)
```

```python
# Countplot for categorical columns
cat_col = ['term', 'grade', 'employment_length', 'purpose', 'application_type']

sns.set_theme()
fig, ax = plt.subplots(nrows=len(cat_col), ncols=2, figsize=(15, 50))
fig.subplots_adjust(hspace=.5)
plt.title('Countplot for categorical variables')

for i, col in enumerate(cat_col):
    sns.countplot(y=col, data=X_valid, order=sorted(list(X_valid[col].unique())), ax = ax[i][0])
    sns.countplot(y=col, data=test, order=sorted(list(test[col].unique())), ax = ax[i][1])
    ax[i][0].set_title('count of {}'.format(col))
    ax[i][1].set_title('count of {}'.format(col))
```
